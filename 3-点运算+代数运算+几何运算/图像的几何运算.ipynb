```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# 直方图计算函数
def GetHist(img):
    img = img.flatten()
    img = img.tolist()
    hist = []
    for i in range(0, 256):
        hist.append(img.count(i))
    return hist
```

# 3. 图像的基本运算

## 3.3 图像的几何运算

### 3.3.1 图像的平移

```python
image = cv2.imread('./png_files/1.png')
h, w = image.shape[:2]

# 定义平移矩阵，需要是numpy的float32类型
# x轴平移200，y轴平移100, 2*3矩阵
M = np.array([[1, 0, 50], 
              [0, 1, 80]], dtype=np.float32)

# 用仿射变换实现平移
new_image = cv2.warpAffine(image, M, (h, w), borderValue=(0, 0, 0))

plt.figure()
plt.subplot(121)
plt.imshow(image)
plt.subplot(122)
plt.imshow(new_image)
plt.show()
```

### 3.3.2 图像的镜像

```python
image = cv2.imread('./png_files/1.png')

# Flipped Horizontally 水平翻转
Horizontal = cv2.flip(image, 1)

# Flipped Vertically 垂直翻转
Vertical = cv2.flip(image, 0)

plt.figure()
plt.subplot(131)
plt.imshow(image)
plt.subplot(132)
plt.imshow(Horizontal)
plt.subplot(133)
plt.imshow(Vertical)
plt.show()
```

### 3.3.3 图像的旋转

```python
image = cv2.imread('./png_files/1.png')
h, w = image.shape[:2]

Beta = np.pi/6

# 定义旋转矩阵
M = np.array([[ np.cos(Beta), np.sin(Beta), 0], 
              [-np.sin(Beta), np.cos(Beta), 0]], dtype=np.float32)

# 用仿射变换实现旋转
new_image = cv2.warpAffine(image, M, (h, w))

plt.figure()
plt.subplot(121)
plt.imshow(image)
plt.subplot(122)
plt.imshow(new_image)
plt.show()
```

### 3.3.4 图像的缩放

```python
matrix = [[255,0,255],
          [0,255,0 ],
          [255,255,0]]

image = np.array(matrix, dtype=np.uint8)


new_image = cv2.resize(image, (0, 0), fx=3, fy=3, interpolation=cv2.INTER_NEAREST)
print(new_image)

plt.figure()
plt.subplot(121)
plt.imshow(image, cmap='gray')
plt.subplot(122)
plt.imshow(new_image, cmap='gray')
plt.show()
```
